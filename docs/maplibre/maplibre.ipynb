```python
from anymap_ts import Map

m = Map(center=[-122.4, 37.8], zoom=10)
m.add_basemap("OpenStreetMap")
m.add_draw_control(position="top-left")
m.add_layer_control()
m
```

```python
m.get_draw_data()
```

```python
m.to_html("map.html")
```
