# WMS Layer Example

Web Map Service integration using deck.gl WMSLayer.

```python
from anymap_ts import DeckGLMap
```

## Create a map with WMS layer

```python
m = DeckGLMap(center=[-95, 40], zoom=4)
m.add_basemap("CartoDB.DarkMatter")

# Add NOAA Weather Radar WMS
m.add_wms_layer(
    data="https://mesonet.agron.iastate.edu/cgi-bin/wms/nexrad/n0r.cgi",
    service_type="wms",
    layers=["nexrad-n0r"],
    opacity=0.8,
)
m
```

## Export to HTML

```python
m.to_html("wms_layer_example.html")
```
