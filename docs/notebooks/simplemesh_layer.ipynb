# Simple Mesh Layer Example

3D mesh geometry rendering using deck.gl SimpleMeshLayer.

```python
from anymap_ts import DeckGLMap
```

## Define mesh data positions

```python
# Sample mesh data positions (locations for mesh instances)
mesh_data = [
    {"position": [-122.4, 37.8], "color": [255, 140, 0, 255]},
    {"position": [-122.38, 37.79], "color": [0, 200, 255, 255]},
    {"position": [-122.42, 37.78], "color": [255, 100, 100, 255]},
    {"position": [-122.39, 37.81], "color": [100, 255, 100, 255]},
    {"position": [-122.41, 37.77], "color": [200, 100, 255, 255]},
]
```

## Create map with simple mesh layer

```python
# Using a GLB model from Khronos glTF samples
MESH_URL = "https://raw.githubusercontent.com/KhronosGroup/glTF-Sample-Models/master/2.0/Duck/glTF-Binary/Duck.glb"

m = DeckGLMap(center=[-122.4, 37.79], zoom=13, pitch=60, bearing=-17)
m.add_basemap("CartoDB.DarkMatter")

m.add_simple_mesh_layer(
    data=mesh_data,
    mesh=MESH_URL,
    get_position="position",
    get_color="color",
    size_scale=500,
    pickable=True,
)
m
```

## Export to HTML

```python
m.to_html("simplemesh_layer_example.html")
```
