# 3D Tiles Layer Example

OGC 3D Tiles visualization using deck.gl Tile3DLayer.

```python
from anymap_ts import DeckGLMap
```

## Create a map with 3D Tiles

```python
m = DeckGLMap(center=[-75.152408, 39.946975], zoom=16, pitch=60, bearing=-17)
m.add_basemap("CartoDB.DarkMatter")

# Add a 3D Tiles dataset
m.add_tile3d_layer(
    data="https://assets.ion.cesium.com/43978/tileset.json",
    opacity=1.0,
    point_size=2,
)
m
```

## Export to HTML

```python
m.to_html("tile3d_layer_example.html")
```
