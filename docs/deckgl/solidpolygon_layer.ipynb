# Solid Polygon Layer Example

Filled polygon visualization using deck.gl SolidPolygonLayer.

```python
from anymap_ts import DeckGLMap
```

## Define polygon data (building footprints)

```python
polygons = [
    {
        "polygon": [
            [-122.402, 37.795],
            [-122.400, 37.795],
            [-122.400, 37.793],
            [-122.402, 37.793],
        ],
        "elevation": 200,
        "color": [255, 140, 0, 200],
    },
    {
        "polygon": [
            [-122.398, 37.796],
            [-122.395, 37.796],
            [-122.395, 37.793],
            [-122.398, 37.793],
        ],
        "elevation": 350,
        "color": [0, 200, 255, 200],
    },
    {
        "polygon": [
            [-122.405, 37.790],
            [-122.401, 37.790],
            [-122.401, 37.787],
            [-122.405, 37.787],
        ],
        "elevation": 150,
        "color": [255, 100, 100, 200],
    },
]
```

## Create map with solid polygon layer

```python
m = DeckGLMap(center=[-122.4, 37.79], zoom=14, pitch=45, bearing=-17)
m.add_basemap("CartoDB.DarkMatter")

m.add_solid_polygon_layer(
    data=polygons,
    get_polygon="polygon",
    get_elevation="elevation",
    get_fill_color="color",
    extruded=True,
    elevation_scale=10,
    pickable=True,
)
m
```

## Export to HTML

```python
m.to_html("solidpolygon_layer_example.html")
```
